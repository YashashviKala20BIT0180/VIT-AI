```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("drug200.csv", encoding='cp1252')
df.head()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = df.dropna()
```

```python
x = df.drop("Drug", axis=1)
y = df["Drug"]
x.head()
```

```python
y.head()
```

```python
label_encoder = LabelEncoder()
x["Sex"] = label_encoder.fit_transform(x["Sex"])
#F is 0 and M is 1
```

```python
x.head()
```

```python
x["BP"] = label_encoder.fit_transform(x["BP"])
```

```python
x.head()
```

```python
x["Cholesterol"] = label_encoder.fit_transform(x["Cholesterol"])
```

```python
x.head(10)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x.head()
```

```python
y_encoded = label_encoder.fit_transform(y)
```

```python
y_encoded
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x_scaled,y_encoded,test_size=0.2,random_state=12)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=x_scaled.shape[1]))
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(5, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,batch_size=4,epochs=50)
```

```python
print(model.summary())
```

```python
random_samples = np.random.randn(5, xtest.shape[1])
random_samples
```

```python
predictions = model.predict(random_samples)
predicted_labels = np.argmax(predictions, axis=1)
```

```python
predicted_drugs = label_encoder.inverse_transform(predicted_labels)
```
